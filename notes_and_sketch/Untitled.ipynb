```python
class LinkedList:
    def __init__(self):
        self.head = None

    def find(self, value):
		# start at the head
        cur = self.head
        while cur is not None:
            if cur.value == value:
                return cur.value
            cur = cur.next
        return None # we didn't find it

# # (11) -> (22) -> (33) -> (44) -> None

# # we need to insert a value and change the pointers .next

    def insert_at_head(self, node):
        n = node
        # new value .next is current head
        n.next = self.head
        # current head is now set to n
        self.head = n

# 	# stretch goal:
# 	def append(self, value):
# 		pass

# # (11) -> (22) -> (33) -> (44) -> None
# 	#prev 	#cur

    def delete(self, value):

        cur = self.head

#Special Case of Deleting Head:
        if cur.value == value: # are we deleting the head?
            self.head = self.head.next
            return cur

# General Case
        prev = cur
        cur = cur.next
        while cur is not None:
            if cur.value == value:
                prev.next = cur.next # cuts out the node
                return cur
            else:
                prev = prev.next
                cur = cur.next

        return None
```
